### Images are different lengths and sizes, and are npt gray.

```python
import os
from PIL import Image

input_dir = 'data/Chickenpox_original'

output_dir = 'converted_data/Chickenpox_grayscale'


if not os.path.exists(output_dir):
    os.makedirs(output_dir)

# loop through each file in the input directory
for filename in os.listdir(input_dir):
    # check if the file ends with .jpeg (case-insensitive)
    if filename.lower().endswith('.jpeg'):
        # construct the full file path
        file_path = os.path.join(input_dir, filename)
        
        # open the image
        img = Image.open(file_path)
        # convert the image to grayscale ('L' mode)
        gray_img = img.convert('L')
        
        # change the extension from .jpeg to .jpg while keeping the same base name
        new_filename = os.path.splitext(filename)[0] + '.jpg'
        # construct the output file path
        output_path = os.path.join(output_dir, new_filename)
        
        # cave the converted image in the output directory
        gray_img.save(output_path)
        print(f"Converted {filename} to {new_filename}")
```

## Converting to gray scale

```python
# overwriting the coloured files
import os
from PIL import Image

folder_path = "color"

for file_name in os.listdir(folder_path):
    if file_name.lower().endswith(".jpg"):
        file_path = os.path.join(folder_path, file_name)
        img = Image.open(file_path)
        gray_img = img.convert("L")
        gray_img.save(file_path)
        print(f"Converted {file_name} to grayscale.")
```

### Checking the numbers

```python
folders = [
    "converted_data/Chickenpox_grayscale",
    "converted_data/Monkeypox_gray"
]

image_extensions = (".jpg")

# iterate over the folders and count the images.
for folder in folders:
    try:
        count = sum(1 for file in os.listdir(folder) if file.lower().endswith(image_extensions))
        print(f"There are {count} images in folder {folder}.")
    except FileNotFoundError:
        print(f"Folder not found: {folder}")
```

```python
from PIL import Image

# step 1: Convert from .jpeg to .jpg
input_file = "other_data/chicken_pox.jpeg" 
jpg_file = os.path.splitext(input_file)[0] + ".jpg"

# open the .jpeg image and save it as a .jpg file
img = Image.open(input_file)
img.save(jpg_file)
print(f"Converted {input_file} to {jpg_file}.")

# step 2: Convert the .jpg image to grayscale
gray_file = os.path.splitext(jpg_file)[0] + "_gray.jpg"
img_jpg = Image.open(jpg_file)
gray_img = img_jpg.convert("L")
gray_img.save(gray_file)
print(f"Converted {jpg_file} to grayscale and saved as {gray_file}.")

```
